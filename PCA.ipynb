El método de PCA permite por lo tanto "condensar" la información aportada por múltiples variables en solo unas pocas componentes. Aun así, no hay que olvidar que sigue siendo necesario disponer del valor de las variables originales para calcular las componentes. Dos de las principales aplicaciones del PCA son la visualización y el preprocesado de predictores previo ajuste de modelos supervisados.

```python
# Tratamiento de datos
# ==============================================================================
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Gráficos
# ==============================================================================
import matplotlib.pyplot as plt
import matplotlib.font_manager
from matplotlib import style
style.use('ggplot') or plt.style.use('ggplot')

# Preprocesado y modelado
# ==============================================================================
from sklearn.decomposition import PCA
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import scale

# Configuración warnings
# ==============================================================================
import warnings
warnings.filterwarnings('ignore')
```

```python
USArrests = sm.datasets.get_rdataset("USArrests", "datasets")
datos = USArrests.data
```

# Exploración inicial


Los dos principales aspectos a tener en cuenta cuando se quiere realizar un PCA es identificar el valor promedio y dispersión de las variables.

La media de las variables muestra que hay tres veces más secuestros que asesinatos y 8 veces más asaltos que secuestros.

```python
print('----------------------')
print('Media de cada variable')
print('----------------------')
datos.mean(axis=0)
```

```python
print('-------------------------')
print('Varianza de cada variable')
print('-------------------------')
datos.var(axis=0)
```

```python
# Entrenamiento modelo PCA con escalado de los datos
# ==============================================================================
pca_pipe = make_pipeline(StandardScaler(), PCA())
pca_pipe.fit(datos)

# Se extrae el modelo entrenado del pipeline
modelo_pca = pca_pipe.named_steps['pca']
```

# Ejemplo pagina ciencia de datos


```python
x1 = [2.5, 0.5, 2.2, 1.9, 3.1, 2.3, 2, 1, 1.5, 1.1]
x2 = [2.4, 0.7, 2.9, 2.2, 3.0, 2.7, 1.6, 1.1, 1.6, 0.9]
```

```python
datos = pd.DataFrame(list(zip(x1,x2))).rename(columns={0:'x1',1:'x2'})
datos
```

```python
mean_x1 = np.mean(datos['x1'])
mean_x2 = np.mean(datos['x2'])
```

```python
datoscentrados = pd.DataFrame(list(zip(datos['x1'] - mean_x1,datos['x2'] - mean_x2))).rename(columns={0:'x1-mean',1:'x2-mean'})
```

```python
plt.scatter(datos['x1'],datos['x2'])
plt.axhline(y=mean_x2, color='r', linestyle='-')
plt.axvline(x=mean_x1, color = 'b', linestyle = '-')
plt.show()
```

```python
plt.scatter(datoscentrados['x1-mean'], datoscentrados['x2-mean'])
plt.axhline(y=mean_x2, color='r', linestyle='-')
plt.axvline(x=mean_x1, color = 'b', linestyle = '-')
plt.show()
```

```python
mean_x1_cen = np.mean(datoscentrados['x1-mean'])
mean_x2_cen = np.mean(datoscentrados['x2-mean'])
```

```python
plt.scatter(datoscentrados['x1-mean'], datoscentrados['x2-mean'])
plt.axhline(y=mean_x1_cen, color='r', linestyle='-')
plt.axvline(x=mean_x2_cen, color = 'b', linestyle = '-')
plt.show()
```

```python
# calculamos la covarianza
covx1x2 = np.cov(datoscentrados['x1-mean'],datoscentrados['x2-mean'])[0][1]
Mcovx1x2 = np.cov(datoscentrados['x1-mean'],datoscentrados['x2-mean'])
covx1x2
```

```python
Mcovx1x2
```

```python
eigenvalue , eigenvector = np.linalg.eig(Mcovx1x2)
```

```python
eigenvalue
```

```python
eigenvector_t = eigenvector.T
```

```python
datoscentrados_t = datoscentrados.T
datoscentrados_t
```

```python
pca_component = pd.DataFrame(np.dot(pd.DataFrame(eigenvector_t), datoscentrados_t)).T
pca_component
```

```python
plt.scatter(pca_component[0], pca_component[1])
plt.show()
```
